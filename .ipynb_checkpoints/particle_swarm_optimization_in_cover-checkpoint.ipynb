```python
#coding:utf-8

from __future__ import division
import random
import math
import os
import pandas as pd
import numpy as np
from numpy.random import randint
from time import time
from numpy import zeros
import matplotlib.pyplot as plt
from numpy.random import randint, uniform
from logging import basicConfig, INFO, debug, info,warn,error, critical


basicConfig(
    # filename="pso_log.log", 
    level= INFO,
    format = '%(asctime)s  %(levelname)s [%(name)s:%(lineno)s] :  %(message)s',
)

coor_x_left = 742670
coor_x_right = 746790
coor_y_left = 3851730
coor_y_right = 3847650
WeakCovergeValue = -110
OverCoverValue = -70
coor_x = {}
i = 0
for x in range(coor_x_left,coor_x_right+1, 20):
    coor_x[x] = i
    i = i + 1
coor_y = {}
i = 0
for y in range(coor_y_left,coor_y_right-1, -20):
    coor_y[y] = i
    i = i + 1
all_id_list = []
#get the dict of filename associate with its parameters from the destinated file
def get_file_names_with_para(para_data_file):
    para_dict = {}
    with open(para_data_file, 'rb') as f:
        df_csv = pd.read_excel(f)
    for i in range(len(df_csv)):
        para_dict[df_csv.iloc[i][0]] = [df_csv.iloc[i][4], df_csv.iloc[i][5], df_csv.iloc[i][6]]
        all_id_list.append(df_csv.iloc[i][0])
    return para_dict

#constitute a mat from a selected file name
def get_mat_from_file(filepath):
    mat =  np.zeros((len(coor_x),len(coor_y)))
    with open(filepath) as f:
        df_csv = pd.read_csv(f)
    for index, line in df_csv.iterrows():
        mat[coor_x[line[3]]][coor_y[line[4]]] = line[7]
    return mat

#get file-names from destinated directory "path"
def read_files(path):
    files = os.listdir(path)
    files.sort()
    s = []
    for file in files:
        if file[-4:] == ".csv":
            s.append(file)
    return s

#get matrix list from files
def get_mats(chosed_site_names):
    all_mats = []
    for site in chosed_site_names:
        all_mats.append(get_mat_from_file(site))
    return all_mats

#load the whole files to mats
def get_mat_dict(site_ids,data_dir):
    whole_mats_dict = {}
    for id in site_ids:
        name = data_dir + "/output" + str(id) + ".csv"
        whole_mats_dict[id] = get_mat_from_file(name)
        print("output" + str(id) + ".csv load finished")
    return whole_mats_dict

#calculate object-function from mats
def calculate_obj(mats):
    result_mat = np.zeros((len(coor_x),len(coor_y)))
    for i in range(len(mats[0])):
        for j in range(len(mats[0][0])):
            over3 = 0
            numOflow = 0
            for mat in mats:
                if abs(mat[i][j]) < 0.001 or (mat[i][j] < WeakCovergeValue):   #judge weak cover
                    numOflow += 1 
                if abs(mat[i][j]) > 0.001 and mat[i][j] > OverCoverValue :  #judge over cover
                    over3 += 1
            if over3 >= 3:
                result_mat[i][j] = 2     #means over cover
                break
            if numOflow == len(mats):   
                result_mat[i][j] = 1      #means weak cover
                break
    #print(result_mat)
    weak_girds_num = 0
    over_girds_num = 0
    for  i in range(len(result_mat)):
        for j in range(len(result_mat[0])):
            if result_mat[i][j] == 1:
                weak_girds_num = weak_girds_num + 1
            if result_mat[i][j] == 2:
                over_girds_num = over_girds_num + 1
    all_girds = len(result_mat) * len(result_mat[0])
    weakCoveragePercentage = float(weak_girds_num / all_girds)
    overCoveragePercentage = float(over_girds_num / all_girds)
    return weakCoveragePercentage,overCoveragePercentage

# ----------------------程序执行-----------------------+
# if __name__ == '__main__':
    #path = "D:/OneDrive/cover-algorithm-code/"  #csv file directory

para_data_file = "/Users/qimeihu/Desktop/cover-algorithm-code/zhengzhou_yanzheng/para.xlsx"
data_dir_path = "/".join(para_data_file.split("/")[:-1])
# data_dir_path = os.pat.split(para_data_file)[0]
para_dict = get_file_names_with_para(para_data_file)
all_file_ids = []
for id in para_dict.keys():
    all_file_ids.append(id)
print("********************* load files started ********************")
t1 = time()
whole_mats_dict = get_mat_dict(all_file_ids,data_dir_path)
print("********************* load files finished ********************")
t = time() - t1
print("used {0}min{1}s to load {2} files".format(int(t/60), int(t%60), len(all_file_ids)))
# print("used {0} seconds to load {1} files"(t, len(all_file_ids)))
```

```python
# ----------------------PSO---------------------------------
class PSO():
    """the primary class of this algrothem to solve
    cover prblem
    """
    def __init__(self, pN, BSnum, max_iter, para_dict, whole_mats_dict):
        """initialize the PSO with 
        """
        self.para_dict = para_dict
        self.whole_mats_dict = whole_mats_dict
        self.BSnum = BSnum
        self.weighOfOverCov = 0.5
        self.weakList = []
        self.overList = []
        self.w = 0.8
        self.c1 = 2
        self.c2 = 2
        self.r1 = 0.6
        self.r2 = 0.3
        self.pN = pN  # 粒子数量
        self.dim = BSnum * 3  # 搜索维度
        self.max_iter = max_iter  # 迭代次数
        self.X = np.zeros((self.pN, self.dim))  # 所有粒子的位置
        self.V = np.zeros((self.pN, self.dim))  # 所有例子的速度
        self.pbest = np.zeros((self.pN, self.dim))  # 个体经历的最佳位置
        self.gbest = np.zeros((1, self.dim))  # 全局最佳位置
        self.p_fit = np.zeros(self.pN)  # 每个个体的历史最佳适应值
        
        self.gfit = 0.5  # 全局最佳适应值
        self.fit_list = []

    def init_Population(self):
        #TODO: never use difinate number 3, change it to variable!
        """initialize the poplulation """
        self.swarm = []              
        for i in range(self.pN):
            particle_id = []
            for x in range(0,len(all_file_ids),9):
                particle_id.append(all_file_ids[randint(x,x+9)])
            self.swarm.append(particle_id)
            
        for i in range(self.pN):
            for j in range(0, self.dim, 3):
                id = self.swarm[i][int(j/3)]
                self.X[i][j] = self.para_dict[id][0] # height
                self.V[i][j] = self.para_dict[id][0]
                self.X[i][j+1] = self.para_dict[id][1]  # dip_angle
                self.V[i][j+1] = self.para_dict[id][1]
                self.X[i][j+2] = self.para_dict[id][2]  # direction
                self.V[i][j+2] = self.para_dict[id][2] 
            self.pbest[i] = self.X[i]
            tmp = self.evaluate(self.swarm[i])
            self.p_fit[i] = tmp
            if tmp < self.gfit:
                self.gfit = tmp          #全局最佳适应值
                self.fit_list.append(tmp)
                self.gbest = self.X[i]  #全局最优个体
#             print("发生更改！\n全局最佳适应值：\n",self.fit,"全局最优个体：\n",self.gbest)

# ---------------------fitness函数-----------------------------
    def evaluate(self, particle_ids_list):
        """evaluate the fitness of a particle
        using the over-cover-rate and weak-cover-rate
        """
        mats = []
        for particle_id in particle_ids_list:
            mats.append(whole_mats_dict[particle_id])
        result_mat = np.zeros((len(coor_x),len(coor_y)))
        for i in range(len(mats[0])):
            for j in range(len(mats[0][0])):
                over3 = 0
                numOflow = 0
                for mat in mats:
                    if abs(mat[i][j]) < 0.001 or (mat[i][j] < WeakCovergeValue):   #judge weak cover
                        numOflow += 1
                        if numOflow == len(mats):   
                            result_mat[i][j] = 1      #means weak cover
                    elif abs(mat[i][j]) > 0.001 and mat[i][j] > OverCoverValue :  #judge over cover
                        over3 += 1
                        if over3 >= 3:
                            result_mat[i][j] = 2     #means over cover
        weak_girds_num = 0
        over_girds_num = 0
        for  i in range(len(result_mat)):
            for j in range(len(result_mat[0])):
                if result_mat[i][j] == 1:
                    weak_girds_num = weak_girds_num + 1
                if result_mat[i][j] == 2:
                    over_girds_num = over_girds_num + 1
        all_girds = len(result_mat) * len(result_mat[0])
        weakCoveragePercentage = float(weak_girds_num / all_girds)
        overCoveragePercentage = float(over_girds_num / all_girds)
        fitness = self.weighOfOverCov *  overCoveragePercentage + (1 - self.weighOfOverCov) * weakCoveragePercentage
#         print("end 1 time",time()-t0)
        
#         self.weakList.append(weakCoveragePercentage)
#         self.overList.append(overCoveragePercentage)
#         print("weak girds:",weak_girds_num," in ",all_girds)
#         print("over girds:",over_girds_num," in ",all_girds)
#         print("fitness = ", fitness)
        return fitness

# ----------------------迭代，更新粒子位置----------------------------------
    def iterator(self):
        """iterate N times
        """
        self.fitness = []
        for t in range(self.max_iter):
            for i in range(0,self.pN):  # 更新gbest\pbest
                self.temp = self.evaluate(self.swarm[i])
                if(self.temp < self.p_fit[i]):  # 更新个体最优
                    self.p_fit[i] = self.temp
                    self.pbest[i] = self.X[i]
#                     print("更新个体最优：\n 个体最佳适应值：",self.p_fit[i],"个体经历的最佳位置：",self.pbest[i])
                    if(self.p_fit[i] < self.gfit):  # 更新全局最优
                        self.gbest = self.X[i]
                        self.fit_list.append(self.p_fit[i])
                        self.gbest_ids = self.swarm[i]
                        self.gfit = self.p_fit[i]
                        print("更新全局最优：\n全局最佳适应值：",self.gfit,"全局最优个体：",self.gbest)
                        print("gbest_ids = ", self.gbest_ids)
            for i in range(0,self.pN):
                self.V[i] = self.w * self.V[i] + self.c1 * self.r1 * (self.pbest[i] - self.X[i]) + \
                    self.c2 * self.r2 * (self.gbest - self.X[i])
                self.X[i] = self.X[i] + self.V[i]

                # 将更新的建议取值改为最接近的现有取值组合
                for j in range(0, len(self.X[0]),3):
                    height,angle,ditrction = self.X[i][j], self.X[i][j+1],  self.X[i][j+2]
                    differ = {}
                    for num in range(0,9):
                        id = all_id_list[int(j/3)*9+num]
                        differ[id] = abs(height - self.para_dict[id][0])/30 + abs(angle - self.para_dict[id][1])/4 + abs(ditrction - self.para_dict[id][2])/180
                    d = sorted(differ.items(),key=lambda item:item[1])
                    selectid = d[0][0]
                    self.swarm[i][int(j/3)] = selectid
                    self.X[i][j] = para_dict[selectid][0]
                    self.X[i][j+1] = para_dict[selectid][1]
                    self.X[i][j+2] = para_dict[selectid][2]
                    
            self.fitness.append(self.temp)
            print("the",t+1,"th iterate finished,gfit = ",self.gfit, "gbst：", self.gbest)
#         print("gbest:",self.gbest)
#         print("fit:",self.fit)  # 输出最优值

#---------------------main-------------------------+
iter = 100
print("********************* start particle swarm optimization *****************")
starttime = time()
my_pso = PSO(pN=30, BSnum=9, max_iter=iter, para_dict=para_dict, whole_mats_dict=whole_mats_dict)
my_pso.init_Population()
fitness = my_pso.iterator()
print("********************* end particle swarm optimization *****************")
endtime = time()
totaltime = endtime - starttime
print("pso finished,using {0}min{1}s".format(int(totaltime/60), int(totaltime%60)))
```
